```python
import pyBigWig

# Open the BigWig file
ctcf_bw_1 = pyBigWig.open('/mnt/home/tudomlumleart/ceph/05_Sox9Dataset/chip_seq_paper/GSM6940137_H9_CTCF_RPGC.bw')
ctcf_bw_2 = pyBigWig.open('/mnt/home/tudomlumleart/ceph/05_Sox9Dataset/chip_seq_paper/GSM6940138_CR16_CTCF_RPGC.bw')
rad21_bw_1 = pyBigWig.open('/mnt/home/tudomlumleart/ceph/05_Sox9Dataset/chip_seq_paper/GSM6940143_H9_Rad21_RPGC.bw')
rad21_bw_2 = pyBigWig.open('/mnt/home/tudomlumleart/ceph/05_Sox9Dataset/chip_seq_paper/GSM6940144_CR16_Rad21_RPGC.bw')

data_list = [ctcf_bw_1, ctcf_bw_2, rad21_bw_1, rad21_bw_2]
```

```python
chrom = 'chr17'
start = 68416487
end = 70868424

summary_list = []
for data in data_list:
    summary = data.stats(chrom, start, end, type="mean", nBins=2000)
    summary_list.append(summary)
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
for i, summary in enumerate(summary_list):
    ax[i // 2, i % 2].plot(summary)
    ax[i // 2, i % 2].set_title(f'Sample {i + 1}')
```
